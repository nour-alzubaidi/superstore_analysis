```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_excel(r"C:\Users\user\OneDrive\Desktop\sales analysis and dashboard\data\clean_data.xlsx")
```

```python
sns.histplot(df['profit'], bins=50)
plt.title('Profit Distribution')
plt.show()
```

#### most profits are small or near zero, most orders generate small profit and there is a low variability in the profit

```python
sns.scatterplot(x= 'discount', y='profit', data=df)
plt.title('Discount vs Profit')
plt.show()
```

#### Higher discounts are associated with lower profit and increased losses

```python
sns.boxplot(x='category', y='profit', data=df,
           
           medianprops=dict(color="red", linewidth=2),
           capprops=dict(color="black", linewidth=2)
)
plt.ylim(-900, 900)  
plt.title('Profit Distribution by Category')
plt.show()
```

#### The furniture category shows higher variability and more frequent losses compared to other categories

```python
monthly_sales = df.resample('MS', on='order_date')['sales'].sum()
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index, monthly_sales.values, color='teal', linewidth=2)
plt.title('Monthly Sales Trend', fontsize=18, weight='bold')
plt.xlabel('Date', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Sales exhibit an upward trend with recurring seasonal peak towards the end of each year

```python
corr = df[['sales', 'profit', 'discount', 'quantity']].corr()
sns.heatmap(corr, annot=True)
plt.title('Correlation Matrix')
plt.show()
```

#### Discount has a negative correlation with profit, confirming earlier findings
